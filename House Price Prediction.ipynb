```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split             
#many datasets are present in sklearn.datasets readily available to use
# train_test_split splits the data into partitions for train and test
from sklearn.linear_model import LinearRegression       #all the linear regression formula are present in this package
from sklearn import metrics                             #used to calculate the accuracy of data
```

```python
house_price_dataset = sklearn.datasets.load_boston()       #dataset for predicting house price for boston
```

```python
house_price_dataset               #now this dataset is not in a readable for so we convert it into a dataframe
```

```python
house_df = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
#house_price_dataset.data returns the data from the dataset 
#columsns = feature_names sets the no of column to be equal to no of features there are in the dataset
```

```python
house_df
```

```python
house_price_dataset.DESCR            #describes the house_price_dataset 
```

```python
house_df['Price'] = house_price_dataset.target
```

```python
correlation = house_df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws = {'size':10}, cmap='Blues')

#cbar is the colorbar located on the right side of the graph
#square is the shape of boxes in the graph
#fmt .1f specifies the limit of floating points in the value
#annot is for annotations which displays the value inside the boxes
#annot_kws size is the font size of annotations and cmap is the color of boxes (Blues,Greens,Reds,Oranges,Purples)
```

```python
x = house_df.drop(['Price'], axis=1)         # we have slpit the data in 2 parts, x are independent variables
y = house_df['Price']                        # y is the dependent value which we have to predict
```

```python
x.shape, y.shape
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
# we have split the data into two parts, 80% training data and 20% testing data
#random_state will select a random value from the available 506 values until they are slpit 
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
model = LinearRegression()           #applying linear regression
model.fit(x_train, y_train)          #training the model, giving the values
```

```python
training_data_prediction = model.predict(x_train)         #predicting value from the training data itself
```

```python
print(training_data_prediction)
```

```python
# to measure if the predicted values are correct and what is it's accuracy

R_square_score = metrics.r2_score(y_train, training_data_prediction)                        
# R**2 = 1 - SSR/TSS
# if R square is near to 1 that means the model has performed very good training

Mean_absolute_error = metrics.mean_absolute_error(y_train, training_data_prediction)        
# MAE = (sum(y(hat) - y(i))/n), i.e. summation of difference of predicted data and actual data divided by number of predictions
# MAE is  average difference between predicted and actual data

print('R Squared Error: ',R_square_score)
print('Mean Absolute Error: ', Mean_absolute_error)
```

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_train, training_data_prediction, c='crimson')
p1 = max(max(training_data_prediction), max(y_train))
p2 = min(min(training_data_prediction), min(y_train))
plt.plot([p1,p2], [p1,p2], 'b-')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price')
plt.show()
```
